```python
def find_duplicates(data_list):
    """Returns a dictionary with items and their frequencies."""
    frequency = {}
    for item in data_list:
        frequency[item] = frequency.get(item, 0) + 1
    # Filter to show only items that appeared more than once
    return {k: v for k, v in frequency.items() if v > 1}

print(find_duplicates([1, 2, 2, 3, 'a', 'a', 'b', 'a']))
```

```python
def is_anagram(str1, str2):
    """Checks if two strings are anagrams."""
    # Remove spaces and convert to lowercase for case-insensitive check
    str1_clean = sorted(str1.replace(" ", "").lower())
    str2_clean = sorted(str2.replace(" ", "").lower())
    return str1_clean == str2_clean

print(is_anagram("Listen", "Silent"))
```

```python
def merge_sorted_lists(list1, list2):
    """Merges two sorted lists into one sorted list."""
    merged = []
    i, j = 0, 0
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged.append(list1[i])
            i += 1
        else:
            merged.append(list2[j])
            j += 1
    # Append remaining elements
    merged.extend(list1[i:])
    merged.extend(list2[j:])
    return merged

print(merge_sorted_lists([1, 5, 8], [2, 4, 7, 9]))
```

```python
def filter_and_square(numbers):
    """Squares odd numbers and filters those over 50 using list comprehension."""
    return [
        (x ** 2)
        for x in numbers
        if x % 2 != 0 and x ** 2 <= 50
    ]

print(filter_and_square([1, 2, 5, 7, 10, 11]))
```

```python
def flatten_list(nested_list):
    """Flattens a list of lists into a single list."""
    flat_list = []
    for inner_list in nested_list:
        if isinstance(inner_list, list):
            flat_list.extend(inner_list)
        else:
            flat_list.append(inner_list)
    return flat_list
print(flatten_list([[1, 2], 3, [4, 5, 6]]))
```

```python
def memoized_fib(n, memo={}):
    """Calculates Fibonacci using memoization for efficiency."""
    if n in memo:
        return memo[n]
    if n <= 1:
        return n

    result = memoized_fib(n - 1, memo) + memoized_fib(n - 2, memo)
    memo[n] = result
    return result
print(memoized_fib(10))
```

```python
def count_total_words(file_path):
    """Counts the total number of words in a text file."""
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            # Use split() to count words, handling various whitespace
            word_count = len(content.split())
            return word_count
    except FileNotFoundError:
        return 0

print(f"Total words: {count_total_words('sample.txt')}")
```

```python
def replace_text_in_file(file_path, old_text, new_text):
    """Replaces all occurrences of old_text with new_text in a file."""
    try:
        with open(file_path, 'r') as file:
            content = file.read()

        new_content = content.replace(old_text, new_text)

        with open(file_path, 'w') as file:
            file.write(new_content)

        print(f"Replaced '{old_text}' with '{new_text}' in {file_path}.")
    except FileNotFoundError:
        print(f"Error: File {file_path} not found.")
```

```python
class BankAccount:
    """A basic bank account with deposit and withdrawal."""

    def __init__(self, initial_balance=0):
        # Using a private attribute (by convention)
        self._balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited ${amount}. New balance: ${self._balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self._balance:
            self._balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self._balance}")
            return True
        else:
            print("Insufficient funds or invalid amount.")
            return False

    def get_balance(self):
        return self._balance



```

```python
from datetime import date

def calculate_age(birth_date_str):
    """Calculates age given a birth date string (YYYY-MM-DD)."""
    try:
        birth_year, birth_month, birth_day = map(int, birth_date_str.split('-'))
        birth_date = date(birth_year, birth_month, birth_day)
    except ValueError:
        return "Invalid date format. Use YYYY-MM-DD."

    today = date.today()

    # Calculate initial age
    age = today.year - birth_date.year

    # Adjust age if the birthday hasn't passed this year yet
    if (today.month, today.day) < (birth_date.month, birth_date.day):
        age -= 1

    return age

print(f"Age: {calculate_age('1990-11-20')} years")
```

```python
import random
import string

def generate_password(length=12):
    """Generates a strong, random password."""
    # Define character sets
    lower = string.ascii_lowercase
    upper = string.ascii_uppercase
    digits = string.digits
    symbols = string.punctuation

    # Ensure at least one of each type is included
    password = [
        random.choice(lower),
        random.choice(upper),
        random.choice(digits),
        random.choice(symbols)
    ]

    # Fill the rest of the password length randomly
    all_chars = lower + upper + digits + symbols
    password.extend(random.choice(all_chars) for _ in range(length - 4))

    # Shuffle the final list to randomize the position of the required characters
    random.shuffle(password)

    return "".join(password)
print(f"New Password: {generate_password(16)}")
```

```python
import platform
import os

def get_system_info():
    """Prints basic system information using platform and os modules."""
    print("--- System Information ---")
    print(f"OS/Kernel Name: {platform.system()} {platform.release()}")
    print(f"Architecture: {platform.machine()}")
    print(f"Python Version: {platform.python_version()}")
    try:
        # Get the number of CPU cores
        print(f"CPU Cores (Physical): {os.cpu_count()}")
    except NotImplementedError:
        print("CPU count not available.")

    print(f"Current Working Dir: {os.getcwd()}")

```

```python
def safe_division():
    """Performs division with robust error handling."""
    try:
        num1 = int(input("Enter numerator: "))
        num2 = int(input("Enter denominator: "))

        result = num1 / num2
        print(f"Result: {result}")

    except ValueError:
        # Catches errors if input is not a valid integer (e.g., 'hello')
        print("Error: Invalid input. Please enter only whole numbers.")

    except ZeroDivisionError:
        # Catches the specific error for dividing by zero
        print("Error: Cannot divide by zero.")

    except Exception as e:
        # A catch-all for any other unexpected error
        print(f"An unexpected error occurred: {e}")

    finally:
        # This code runs regardless of whether an exception occurred
        print("Division attempt completed.")
```
